```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import sqlite3
sns.set()
```

```python
conn = sqlite3.connect('autovit.db')
raw_data = pd.read_sql_query('SELECT * FROM autovit_db', conn)

#Remove strings from number fields
raw_data1 = raw_data.replace(['\n', 'km', 'CP', 'cm3'],'', regex=True)

#Remove emptpy values, 'n'
# raw_data1['Year'] = raw_data1['Year'].replace(['n'],'', regex=True)
raw_data1['Mileage'] = raw_data1['Mileage'].replace(['n', '\ '],'', regex=True)
raw_data1['Mileage'] = raw_data1['Mileage'].astype(int)
# raw_data1['Engine_size'] = raw_data1['Engine_size'].replace(['n'],'', regex=True)
# raw_data1['Horsepower'] = raw_data1['Horsepower'].replace(['n'],'', regex=True)
```
